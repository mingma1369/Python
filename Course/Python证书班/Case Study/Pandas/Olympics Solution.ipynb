## Challenge: 
### In every Olympics, which US athlete has won the most total number of medals?

```python
import pandas as pd
```

```python
df = pd.read_csv('olympics.csv',skiprows=4)
df.head()
```

```python
df1  = df[df.NOC == 'USA']
df1.head()
```

```python
df2 = df1.groupby(['Edition','Athlete','Medal']).size()
df2.head()
```

```python
df3 = df2.unstack('Medal',fill_value=0)
df3.head()
```

```python
df3.columns.name = None
df3['Total'] = df3['Gold'] + df3['Silver'] + df3['Bronze']
df3.head()
```

```python
df3.reset_index(inplace=True)
df3.head()
```

```python
df3.sort_values(['Edition','Total'],ascending=[True,False],inplace=True)
df3.head()
```

```python
df3.drop_duplicates(subset='Edition', keep='first', inplace=True)
df3.reset_index(drop=True, inplace=True)
df3
```
