#### Q6. Calculate the total revenue for 2 weeks 

```python
# first combine the two week data together
total_order = pd.concat([week1,week2])
total_order.head()
```

```python
# check if the dataframe size makes sense
len(week1) + len(week2) == len(total_order)
```

```python
# join with food dataframe to get the price
total_order = total_order.merge(food,how = 'inner',on = 'Food ID')
total_order
```

```python
# create the column called revenue
total_order['Revenue'] = total_order['Quantity'] * total_order['Price']
```

```python
total_order['Revenue'].sum()
```

### Homework

### The restaurant wants to do some promotions by sending customers the coupon card. The coupon is 10 dollars, 5 dollars and 2 dollars.  The plan is:
- 2-week total revenue larger than or equal 100, we will give them 10 dollar coupon.
- 2-week total revenue larger than or equal 50, we will give them 5 dollar coupon. 
- 2-week total revenue larger than or equal 10, we will give them 2 dollar coupon. 
- others 0
  
#### Q1.Can you generate the complete list of the coupon assignment?

```python
# Define the function to assign the coupn
def coupon_assign(x):
    if x>= 100:
        return 10
    elif x>=50 and x<100:
        return 5
    elif x>=10 and x<50:
        return 2
    else:
        return 0
```

```python
# since one customer can order multiple times in 2-week period, we need to group by customer ID first to get the total revenue by each person
complete_df_by_customer = complete_df.groupby(['Customer ID','Customer Full Name'])['Revenue'].sum()
complete_df_by_customer = complete_df_by_customer.reset_index()
complete_df_by_customer
```

```python
complete_df_by_customer['Coupon'] = complete_df_by_customer['Revenue'].apply(lambda x : coupon_assign(x))
```
